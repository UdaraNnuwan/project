# 04 GPT Adjudication

This notebook demonstrates the post-threshold GPT decision flow. The FiLM autoencoder remains the anomaly detector; GPT is only applied after threshold crossing for interpretation, false-positive filtering, severity classification, and recommended action generation.


```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

print(f"PROJECT_ROOT: {PROJECT_ROOT}")

import ast
import json
import joblib
import pandas as pd
from IPython.display import display
try:
    from dotenv import load_dotenv
except ImportError:
    def load_dotenv(*args, **kwargs):
        return False

from container_ad_pipeline.config import PipelineConfig, ensure_pipeline_directories
from container_ad_pipeline.dataset import build_dataset_from_csv, load_dataset_bundle
from container_ad_pipeline.evaluate import (
    add_top_k_feature_columns,
    evaluate_predictions,
    inject_synthetic_anomalies,
    run_model_inference,
    save_evaluation_outputs,
)
from container_ad_pipeline.gpt_adjudicator import GPTAdjudicator
from container_ad_pipeline.train import select_split, train_film_autoencoder

load_dotenv(PROJECT_ROOT / ".env")
config = PipelineConfig()
ensure_pipeline_directories(config)
```

## Bootstrap Evaluation Artifacts

This notebook can rebuild the dataset, model, and evaluation outputs if they are not already present, so it remains runnable independently.


```python
if not (config.paths.processed_dir / "X_all.npy").exists():
    build_dataset_from_csv(config.paths.raw_metrics_csv, config.dataset, config.paths.processed_dir)
if not config.paths.checkpoint_path.exists():
    bundle_bootstrap = load_dataset_bundle(config.paths.processed_dir)
    train_film_autoencoder(
        bundle=bundle_bootstrap,
        train_config=config.train,
        checkpoint_path=config.paths.checkpoint_path,
        x_scaler_path=config.paths.x_scaler_path,
        c_scaler_path=config.paths.c_scaler_path,
        detector_meta_json=config.paths.detector_meta_json,
        detector_meta_joblib=config.paths.detector_meta_joblib,
    )
if not (config.paths.evaluation_dir / "window_level_predictions.csv").exists():
    bundle_bootstrap = load_dataset_bundle(config.paths.processed_dir)
    x_scaler = joblib.load(config.paths.x_scaler_path)
    c_scaler = joblib.load(config.paths.c_scaler_path)
    threshold = float(joblib.load(config.paths.detector_meta_joblib)["threshold"])
    X_test, _, _ = select_split(bundle_bootstrap, "test")
    X_injected, y_true, events = inject_synthetic_anomalies(
        X_test,
        feature_columns=bundle_bootstrap.feature_meta["feature_columns"],
        anomaly_ratio=config.eval.anomaly_ratio,
        seed=config.eval.anomaly_seed,
    )
    predictions = run_model_inference(
        bundle=bundle_bootstrap,
        model_path=config.paths.checkpoint_path,
        x_scaler=x_scaler,
        c_scaler=c_scaler,
        split_name="test",
        injected_X=X_injected,
    )
    predictions = add_top_k_feature_columns(predictions, bundle_bootstrap.feature_meta["feature_columns"], config.eval.top_k_features)
    eval_artifacts = evaluate_predictions(predictions, y_true, threshold, config.eval, events)
    save_evaluation_outputs(eval_artifacts, config.paths.evaluation_dir, config.eval)
```

## Step 1. Load Anomalous Windows From Evaluation Results

We begin with the saved evaluation output and keep only threshold-crossing windows. This preserves the design boundary: AE first, GPT second.


```python
bundle = load_dataset_bundle(config.paths.processed_dir)
feature_columns = bundle.feature_meta["feature_columns"]
threshold = float(joblib.load(config.paths.detector_meta_joblib)["threshold"])

eval_results = pd.read_csv(config.paths.evaluation_dir / "window_level_predictions.csv")
eval_results["feature_error_vector"] = eval_results["feature_error_vector"].apply(ast.literal_eval)

anomalous_windows = (
    eval_results[eval_results["predicted_label"] == 1]
    .copy()
    .sort_values("anomaly_score", ascending=False)
    .reset_index(drop=True)
)

print("Threshold:", threshold)
print("Anomalous windows:", len(anomalous_windows))
display(anomalous_windows[["window_id", "namespace", "pod", "container", "anomaly_score"]].head(10))
```

## Step 2. Compute Feature-Wise Reconstruction Errors

The evaluation output already stores a reconstruction-error vector for each window. Here we expand it into explicit per-feature columns.


```python
feature_error_columns = [f"{feature}_recon_error" for feature in feature_columns]
feature_error_df = pd.DataFrame(anomalous_windows["feature_error_vector"].tolist(), columns=feature_error_columns)
anomalous_windows = pd.concat([anomalous_windows, feature_error_df], axis=1)

display(anomalous_windows[["window_id", *feature_error_columns]].head())
```

## Step 3. Select Top-K Anomalous Features

The top-k anomalous features are derived directly from the feature-wise reconstruction errors.


```python
top_k = config.gpt.top_k_features

def select_top_k_features(row: pd.Series) -> pd.Series:
    ranked = sorted(
        [(feature, float(row[f"{feature}_recon_error"])) for feature in feature_columns],
        key=lambda item: item[1],
        reverse=True,
    )
    top_ranked = ranked[:top_k]
    return pd.Series(
        {
            "top_k_features": [feature for feature, _ in top_ranked],
            "top_k_feature_errors": [score for _, score in top_ranked],
        }
    )

top_k_df = anomalous_windows.apply(select_top_k_features, axis=1)
anomalous_windows = pd.concat([anomalous_windows, top_k_df], axis=1)

display(anomalous_windows[["window_id", "anomaly_score", "top_k_features", "top_k_feature_errors"]].head(10))
```

## Step 4. Build A Compact Prompt

Each GPT payload contains the anomaly score, top-k features, a compact context summary, and optional recent logs.


```python
def read_recent_logs(log_path: Path, limit: int = 30) -> list[str]:
    if not log_path.exists():
        return []
    lines = [line.strip() for line in log_path.read_text(encoding="utf-8", errors="ignore").splitlines() if line.strip()]
    return lines[-limit:]

recent_logs = read_recent_logs(PROJECT_ROOT / "product" / "result_dual_status.log")
adjudicator = GPTAdjudicator(config.gpt)

def build_payload(row: pd.Series) -> dict:
    return {
        "anomaly_score": float(row["anomaly_score"]),
        "threshold": float(threshold),
        "normalized_score": float(row["anomaly_score"] / max(threshold, 1e-6)),
        "top_k_anomalous_features": list(row["top_k_features"]),
        "top_k_feature_errors": list(row["top_k_feature_errors"]),
        "context_metadata": {
            "namespace": row["namespace"],
            "pod": row["pod"],
            "container": row["container"],
            "start_time": row["start_time"],
            "end_time": row["end_time"],
        },
        "recent_logs": recent_logs,
        "recent_events": [],
    }

anomalous_windows["payload"] = anomalous_windows.apply(build_payload, axis=1)
anomalous_windows["compact_prompt"] = anomalous_windows["payload"].apply(adjudicator.build_prompt)

print("Loaded recent log lines:", len(recent_logs))
print(anomalous_windows.loc[0, "compact_prompt"][:2000])
```

## Steps 5 And 6. Call GPT And Parse Structured JSON Output

The call is made through the Responses API wrapper in `GPTAdjudicator`. If no API key is present, the notebook falls back to the heuristic adjudicator but keeps the same JSON shape.


```python
parsed_records = []
for _, row in anomalous_windows.head(10).iterrows():
    payload = row["payload"]
    structured_output = adjudicator.adjudicate(payload)
    parsed_records.append(
        {
            "window_id": int(row["window_id"]),
            "namespace": row["namespace"],
            "pod": row["pod"],
            "container": row["container"],
            "anomaly_score": float(row["anomaly_score"]),
            "top_k_features": list(row["top_k_features"]),
            "structured_json": structured_output,
            **structured_output,
        }
    )

gpt_results = pd.DataFrame(parsed_records)
config.paths.adjudication_dir.mkdir(parents=True, exist_ok=True)
gpt_results.to_csv(config.paths.adjudication_dir / "gpt_adjudications.csv", index=False)
(config.paths.adjudication_dir / "gpt_adjudications.json").write_text(
    json.dumps(gpt_results.to_dict(orient="records"), indent=2),
    encoding="utf-8",
)

display(gpt_results[["window_id", "anomaly_score", "label", "severity", "explanation", "recommended_action"]].head(10))
gpt_results.loc[0, "structured_json"]
```

## Step 7. Compare AE-Only Label vs AE+GPT Label

Every threshold-crossing window is an AE-side anomaly candidate. GPT then re-adjudicates that same candidate into `anomaly`, `false_positive`, or `needs_review`.


```python
comparison = gpt_results.copy()
comparison["ae_only_label"] = "anomaly"
comparison["gpt_label"] = comparison["label"]
comparison["gpt_severity"] = comparison["severity"]
comparison["gpt_explanation"] = comparison["explanation"]
comparison["gpt_recommended_action"] = comparison["recommended_action"]

display(comparison[[
    "window_id",
    "namespace",
    "pod",
    "container",
    "anomaly_score",
    "ae_only_label",
    "gpt_label",
    "gpt_severity",
]].head(10))
```
